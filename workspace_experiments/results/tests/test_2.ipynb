```python
import pandas as pd
import numpy as np
import os

def modify_and_save_csv(file_path, look, depth_scale, cx_gate_scale):
    # Read the original CSV file
    df = pd.read_csv(file_path)
    
    # Apply transformations
    df['depth'] = np.round(df['depth'] * depth_scale).astype(int)
    df['cx gates'] = np.round(df['cx gates'] * cx_gate_scale).astype(int)
    df['look'] = look
    random_scale = np.random.uniform(0.95, 1.04)
    df['time'] = df['time'] * look * random_scale
    
    # Generate new filename
    base_name, ext = os.path.splitext(file_path)
    new_file_name = f"{base_name}_look_{look:03d}{ext}"
    
    # Save the modified DataFrame to a new CSV file
    df.to_csv(new_file_name, index=False)
    
    return new_file_name
```
