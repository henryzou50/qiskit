```python


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def compare_csv_scatterplot(filenames, metric, title, x_axis_label, legend_labels):
    # Load the data from each CSV file
    data = [pd.read_csv(filename) for filename in filenames]
    
    # Prepare a figure and axis for plotting
    plt.figure(figsize=(10, 6))
    ax = plt.subplot(111)
    
    # Define markers for each file
    markers = ['o', 's', 'D', '^', 'v', 'p', '*', 'H', '8']

    # Plot each file's metric vs its index with different markers and lines
    for i, df in enumerate(data):
        # Plotting points with scatter
        ax.scatter((df.index * 10) + 10, df[metric], label=legend_labels[i], alpha=0.7, edgecolors='w', s=50, marker=markers[i])
        
        # Connecting points with a line
        ax.plot((df.index * 10) + 10, df[metric], marker=markers[i], linestyle='-', alpha=0.7)
    
    # Customize the plot
    ax.set_title(title)
    ax.set_xlabel(x_axis_label)
    ax.set_ylabel(metric)
    ax.legend()

    # Add grid lines
    ax.grid(True)
    
    # Display the plot
    plt.show()
# Example usage (assuming the file paths are 'data1.csv', 'data2.csv', 'data3.csv'):
# compare_csv_scatterplot('data1.csv', 'data2.csv', 'data3.csv', 'time', 'Comparison of Metric Across Files', 'Row')
```

```python
circuit_name = "zou_circuits/problems_ring_27_sample"
rp1 = 'sabre_v0_20_'
rp2 = 'sabre_v0_20_depth'
rp3 = 'sabre_v0_20_dive_beam_1'
rp4 = 'sabre_v0_20_dive_beam_3'
rp5 = 'sabre_v0_20_dive_beam_5'
rp6 = 'sabre_v0_20_dive_beam_10'
rp7 = 'sabre_v0_20_dive_beam_30'

rp0 = 'optimal_upperbound'
legend_labels = [rp1, rp2, rp3, rp4, rp5, rp6, rp7, rp0]

# renaming for clarity
rp0 = 'None'
rp2 = 'sabre_v0_20_depth_old'


file1 = f'{circuit_name}/{rp1}.csv'
file2 = f'{circuit_name}/{rp2}.csv'
file3 = f'{circuit_name}/{rp3}.csv'
file4 = f'{circuit_name}/{rp4}.csv'
file5 = f'{circuit_name}/{rp5}.csv'
file6 = f'{circuit_name}/{rp6}.csv'
file7 = f'{circuit_name}/{rp7}.csv'
file0 = f'{circuit_name}/{rp0}.csv'


filenames = [file1, file2, file3, file4, file5, file6, file7, file0]

metric  = 'depth'
title   = f'Compilation Output Depth for Synthetic Circuits (Ring 27 qubits)'
x_label = "Number of Layers"


compare_csv_scatterplot(filenames, metric, title, x_label, legend_labels)
```
