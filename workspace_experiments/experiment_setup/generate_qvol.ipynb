This notebook generates and stores the ghz circuits to the `ghz` directory. It also lets the user check and see each circuit. 

```python
from qiskit.qasm2 import dump 
from qiskit.circuit.library import QuantumVolume
from circuits.circuit_files import get_circuits_from_directory
```

# Generate and store circuits

```python
def generate_qv_circuits(num_qubits, num_circuits):
    qc_list = []

    for i in range(1, num_circuits + 1):
        qvol = QuantumVolume(num_qubits, i, seed=i)
        qc = qvol.decompose()
        qc_list.append(qc)

    return qc_list
```

```python
num_qubits = 10
num_circuits = 100
qc_list = generate_qv_circuits(num_qubits, num_circuits)
```

```python
# Store the circuits in a file as qvol_10_001.qasm, qvol_10_002.qasm, ... qvol_10_100.qasm
for i, qc in enumerate(qc_list):
    filename = f"circuits/qvol/qvol_{num_qubits}_{i+1:03d}.qasm"
    dump(qc, filename)
```
