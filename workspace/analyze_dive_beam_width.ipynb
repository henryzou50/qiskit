```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ast
```

```python
def plot_data(filename, y_key, title, x_label, y_label, x_key):
    # Load the CSV file
    df = pd.read_csv(filename)

    df['best_data'] = df['best_data'].apply(lambda x: ast.literal_eval(x))
    df[y_key] = df['best_data'].apply(lambda x: x.get(y_key))

    plt.style.use('ggplot')
    sns.set_theme(style="whitegrid")

    plt.figure(figsize=(10, 6))
    scatter = sns.scatterplot(data=df, x=x_key, y=y_key, palette="viridis", s=100, edgecolor="w", alpha=0.7)

    plt.title(title, fontsize=16)
    plt.xlabel(x_label, fontsize=14)
    plt.ylabel(y_label, fontsize=14)

    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    plt.show()
```

GHZ 25

```python
filename = 'data/ghz_25/sabre_v020_dive.csv'
plot_data(filename, 'depth', 'Sabre Dive: GHZ 25 Qubits (Depth vs Beam Width)', 'beam width', 'depth', 'beam width')
plot_data(filename, 'cx_gates', 'Sabre Dive: GHZ 25 Qubits (Depth vs Beam Width)', 'beam width', 'cx gates', 'beam width')
```
