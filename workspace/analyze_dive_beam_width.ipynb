```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ast
```

```python
def plot_data(filename, y_key, title, x_label, y_label):
    # Load the CSV file
    df = pd.read_csv(filename)

    # Convert the string representation of dictionary to actual dictionary
    df['best_data'] = df['best_data'].apply(lambda x: ast.literal_eval(x))

    # Extract the specified key from the dictionary
    df[y_key] = df['best_data'].apply(lambda x: x.get(y_key))

    # Set the style of the matplotlib
    plt.style.use('ggplot')
    sns.set_theme(style="whitegrid")

    # Create the scatterplot
    plt.figure(figsize=(10, 6))
    scatter = sns.scatterplot(data=df, x='beam width', y=y_key, palette="viridis", s=100, edgecolor="w", alpha=0.7)

    # Set the title and labels
    plt.title(title, fontsize=16)
    plt.xlabel(x_label, fontsize=14)
    plt.ylabel(y_label, fontsize=14)

    # Set the tick labels fontsize
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    # Show the plot
    plt.show()
```

```python
filename = 'data/ghz_25/sabre_v020_dive.csv'
plot_data(filename, 'depth', 'Depth vs Beam Width', 'beam width', 'depth')
```
