# Summary plots

```python
from ulti.graphs import compare_single_metric, analyze_metric
```

### Comparing lookahead results to Sabre 0.20 to Sabre 0.20 Lookahead 1
GHZ Circuit (2 qubits to 127 qubits)

Examining depth

```python
dir = "data/ghz_2-127"
file1 = f'{dir}/sabre_v020.csv'
file2 = f'{dir}/sabre_look_1_v020.csv'
compare_single_metric('depth', file1, file2, "Sabre 0.20 Lookahead 1", "Sabre 0.20", "Compilation Output Depth for GHZ circuit on Sherbrooke", "Number of GHZ Qubits")
```

Examining number of CNOTS

```python
compare_single_metric('cx_gates', file1, file2, "Sabre 0.20 Lookahead 1", "Sabre 0.20", "Compilation Output CNOT Gate count for GHZ circuit on Sherbrooke", "Number of GHZ Qubits")
```

```python
dir = "data/ghz_2-127"
baseline_file = 'sabre_v020.csv'
x_label = 'Number of GHZ Qubits'

metric = 'depth'
title = f'{metric} Analysis Across Circuits'
summary_depth = analyze_metric(dir, metric, baseline_file, title, x_label)
print(summary_depth)

metric = 'time'
title = f'{metric} Analysis Across Circuits'
summary_time = analyze_metric(dir, metric, baseline_file, title, x_label)
print(summary_time)

metric = 'cx_gates'
title = f'{metric} Analysis Across Circuits'
summary_cx_gates = analyze_metric(dir, metric, baseline_file, title, x_label)
print(summary_cx_gates)
```
