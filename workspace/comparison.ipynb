```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ast

```

```python
def create_scatterplot(csv_file, metric, x_label, y_label, title, x_value, baseline=None):

    df = pd.read_csv(csv_file)
    # Convert the string representation of dictionary to dictionary
    df['best_data'] = df['best_data'].apply(ast.literal_eval)

    # Extract the metric and beam values
    df[metric] = df['best_data'].apply(lambda x: x.get(metric))

    sns.set_theme(style="whitegrid")
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x=x_value, y=metric, palette="viridis", s=100, marker='o', edgecolor='black')
    plt.xlabel(x_label, fontsize=12)
    plt.ylabel(y_label, fontsize=12)
    plt.title(title, fontsize=15)

    # If a baseline is provided, draw a horizontal line
    if baseline is not None:
        plt.axhline(y=baseline, color='r', linestyle='--')
        plt.text(df[x_value].max(), baseline + (baseline * 0.01), f'', color='r', va='bottom', ha='right', fontsize=10)

    plt.show()
```

```python
# GHZ circuit, varying beam

csv_file = 'data/summary_ghz/vary_beam.csv'  
x_label = 'Beam Size'    
x_value = 'beam'  
metrics = ['depth', 'depth_ratio', 'time', 'cx_gates', 'num_gates']
baseline_info = [300, 3.0, 0.3077833652496338, 1815, 1816]

# comparing depth
y_label = 'Depth'     
title = f'GHZ Circuit 100 Qubits on 127 Qubit Device ({x_label} vs {y_label})'
idx = 0
metric = metrics[idx]
baseline = baseline_info[idx]  
create_scatterplot(csv_file, metric, x_label, y_label, title, x_value, baseline)

# comparing time
y_label = 'Time'     
title = f'GHZ Circuit 100 Qubits on 127 Qubit Device ({x_label} vs {y_label})'
idx = 2
metric = metrics[idx]
baseline = baseline_info[idx]      
create_scatterplot(csv_file, metric, x_label, y_label, title, x_value, baseline)

# comparing cx_gates
y_label = 'CNOT Gates'
title = f'GHZ Circuit 100 Qubits on 127 Qubit Device ({x_label} vs {y_label})'
idx = 3
metric = metrics[idx]
baseline = baseline_info[idx]
create_scatterplot(csv_file, metric, x_label, y_label, title, x_value, baseline)

# comparing num_gates
y_label = 'Total Gates'
title = f'GHZ Circuit 100 Qubits on 127 Qubit Device ({x_label} vs {y_label})'
idx = 4
metric = metrics[idx]
baseline = baseline_info[idx]
create_scatterplot(csv_file, metric, x_label, y_label, title, x_value, baseline)
```
