This notebook will manually check if the circuits from a given directory are built as intended

```python
from experiments.utils.circuits import directory_to_circuits, get_circuits_depth
from IPython.display import display
```

```python
def print_circuit_info(directory):
    # Obtain the circuits, their filenames and their depths
    circuits, filenames  = directory_to_circuits(directory)

    depths = get_circuits_depth(circuits)

    # Print the results for each circuit and manually check if the results are correct
    for i in range(len(circuits)):
        print(f"{filenames[i]}: {depths[i]}")
    
    return circuits
```

GHZ

```python
directory = "experiments/circuits/ghz/"
circuits = print_circuit_info(directory)
```

```python
# draw the first and last circuit
display(circuits[0].draw(output='mpl'))
display(circuits[-1].draw(output='mpl', fold=-1))
```
