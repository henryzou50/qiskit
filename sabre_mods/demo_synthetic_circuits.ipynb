```python
# Set num_layers and num_qubits
num_layers = 3
num_qubits = 12

# Set coupling_map
from qiskit.transpiler import CouplingMap
cm = CouplingMap.from_ring(num_qubits)

# set seed
seed = 42

layers = [
    ('cx', 1),
    ('swap', 1),
    ('cx', 1),
    ('swap', 1),
    ('cx', 1),
    ('swap', 1),
    ('cx', 1),
    ('swap', 1),
    ('cx', 1),
    ('swap', 1),
    ('cx', 1),
    ('swap', 1)
]
```

```python
qc = SyntheticCircuit(cm, layers, seed, cm)
qc = qc.decompose()
qc.draw(output='mpl')
```
