```python
from experiments.utils.synthetic_circuits import SyntheticCircuit
from qiskit.circuit.library.standard_gates import CXGate, SwapGate, CZGate, CYGate, CHGate
```

```python
# Set num_layers and num_qubits
num_layers = 3
num_qubits = 12

# Set coupling_map
from qiskit.transpiler import CouplingMap
cm = CouplingMap.from_ring(num_qubits)

# set seed
seed = 42

layers = [
    (CXGate(), 1.0),
    (SwapGate(), 1.0),
    (CZGate(), 1.0),
    (SwapGate(), 1.0),
    (CYGate(), 1.0),
    (SwapGate(), 1.0),
    (CHGate(), 1.0),
    (SwapGate(), 1.0),
    (CXGate(), 1.0),
    (SwapGate(), 1.0),
    (CXGate(), 1.0),
    (SwapGate(), 1.0)
]

qc = SyntheticCircuit(cm, layers, seed, barriers=True)
qc = qc.decompose()
qc.draw(output='mpl')

```

```python
qc_scrambled = SyntheticCircuit.scramble_qubits(qc, seed)
qc_scrambled.draw(output='mpl')
```
