```python
from experiments.utils.synthetic_circuits import SyntheticCircuit
```

```python
# Set num_layers and num_qubits
num_layers = 3
num_qubits = 12

# Set coupling_map
from qiskit.transpiler import CouplingMap
cm = CouplingMap.from_ring(num_qubits)

# set seed
seed = 42

layers = [
    ('cx', 1.0),
    ('swap', 1.0),
    ('cx', 1.0),
    ('swap', 1.0),
    ('cx', 1.0),
    ('swap', 1.0),
    ('cx', 1.0),
    ('swap', 1.0),
    ('cx', 1.0),
    ('swap', 1.0),
    ('cx', 1.0),
    ('swap', 1.0)
]

qc = SyntheticCircuit(cm, layers, seed, barriers=True)
qc = qc.decompose()
qc.draw(output='mpl')
```
