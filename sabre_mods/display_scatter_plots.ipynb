```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

def compare_csv_scatterplot(filenames, metric, title, x_axis_label, legend_labels, x_values_list):
    # Load the data from each CSV file
    data = [pd.read_csv(filename) for filename in filenames]
    
    # Prepare a figure and axis for plotting
    plt.figure(figsize=(10, 6))
    ax = plt.subplot(111)
    
    # Define markers for each file
    markers = ['o', 's', 'D', '^', 'v', 'p', '*', 'h', 'H', '8', '<', '>', 'd', 'P', 'X']
    
    # Check if the length of x_values_list matches the number of rows in the dataframes
    for df in data:
        if len(df) != len(x_values_list):
            raise ValueError("Length of x_values_list does not match the number of rows in the dataframes.")
    
    # Plot each file's metric with the provided x_values_list
    for i, df in enumerate(data):
        # Ensure x_values_list matches the number of rows in df
        x_values = x_values_list[:len(df)]
        
        # Plotting points with scatter
        ax.scatter(x_values, df[metric], label=legend_labels[i], alpha=0.7, edgecolors='w', s=50, marker=markers[i])
        
        # Connecting points with a line
        ax.plot(x_values, df[metric], marker=markers[i], linestyle='-', alpha=0.7)
    
    # Customize the plot
    ax.set_title(title)
    ax.set_xlabel(x_axis_label)
    ax.set_ylabel(metric)
    ax.legend()

    # Add grid lines
    ax.grid(True)
    
    # Display the plot
    plt.show()


metric = 'depth'
base_directory = 'experiments/results/'
```

GHZ

```python
# Results for ghz_small
target_directory = base_directory + 'ghz_small/'
x_values_list = [10, 20, 30, 40, 50] 
x_label = 'Number of qubits'
title = 'Depth of GHZ circuit'

# Get all of the .csv files in the target directory
csv_files = [target_directory + f for f in os.listdir(target_directory) if f.endswith('.csv')]
csv_files.sort()

# get the names
legend_labels = [f.split('/')[-1].split('.')[0] for f in csv_files]

indices_included = [0, 1, 2, 3, 4, 5, 6, 7, 16]
# reduces the csv files and legend labels to only the ones in indices_included
csv_files = [csv_files[i] for i in indices_included]
legend_labels = [legend_labels[i] for i in indices_included]


compare_csv_scatterplot(csv_files, metric, title, x_label, legend_labels, x_values_list)
```
