```python
from ulti.pass_manager import build_pm, build_pm_list
from ulti.process_files import file_to_circuit_list, file_to_coupling_map
from qiskit.transpiler.passes.routing.sabre_swap import SabreSwap
from qiskit.transpiler.passes.layout.sabre_layout import SabreLayout
```

```python
qc_list_filename = "circuits/ghz/"
cm_filename = "coupling_maps/heavy_hex_027.txt"
qc_list = file_to_circuit_list(qc_list_filename)
cm = file_to_coupling_map(cm_filename)
```

Test build_pm

```python
pm = build_pm(SabreSwap, SabreLayout, cm)
qc = qc_list[6]

# Before transpilation
qc.draw(output="mpl")
```

```python
# After transpilation
qc_tr = pm.run(qc)
qc_tr.draw(output="mpl", idle_wires=False)
```

Test build_pm_list

```python
num_pm = 5
pm_list = build_pm_list(num_pm, SabreSwap, SabreLayout, cm)
for pm in pm_list:
    qc_tr = pm.run(qc)
    print(qc_tr.depth())
```
