```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_depth_vs_beam(filename, x_label, y_label, title, baseline1, baseline2, baseline3):
    # Read the CSV file
    data = pd.read_csv(filename)
    
    # Create the scatter plot
    plt.figure(figsize=(10, 8))
    plt.scatter(data['num_iter'], data['depth'], color='blue', alpha=0.5)
    
    # Add the baselines
    plt.axhline(y=baseline1, color='r', linestyle='--', label=f'Sabre 0.20 (Basic) {baseline1}')
    plt.axhline(y=baseline2, color='g', linestyle='--', label=f'Sabre 0.20 (Extended) {baseline2}')
    plt.axhline(y=baseline3, color='m', linestyle='--', label=f'Sabre 0.20 (Depth) {baseline3}')
    
    # Labeling the axes and the plot
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend()
    
    # Make the plot neat and visually appealing
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.tight_layout()
    
    # Show the plot
    plt.show()
```

```python
filename = 'sabre_025_dive_.csv'  # Replace with the path to your CSV file
x_label = 'Number of Iterations'
y_label = 'Depth'
title = 'QVOL Circuit 10 Qubits'
baseline1 = 1173  # Example baseline value
baseline2 = 1045  # Another example baseline value
baseline3 = 1024

plot_depth_vs_beam(filename, x_label, y_label, title, baseline1, baseline2, baseline3)
```
