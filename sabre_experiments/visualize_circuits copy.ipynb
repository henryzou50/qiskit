```python
from qiskit import QuantumCircuit
from qiskit.circuit.library import QuantumVolume
```

```python
# Create a simple Quantum Volume circuit
qc = QuantumVolume(5, 10, seed=10)
qc.draw(output='mpl')
```

```python
qc_decomposed = qc.decompose() 
qc_decomposed.draw(output='mpl')
```

```python
from qiskit.transpiler.passes.routing.sabre_depth_mirror import SabreSwap
from qiskit.transpiler.passes.layout.trivial_layout import TrivialLayout
from qiskit.transpiler.coupling import CouplingMap
from ulti.pass_manager import build_pm
```

```python
cm = CouplingMap.from_ring(5)
routing_pass = SabreSwap(cm, seed=42)
layout_pass = TrivialLayout(cm)
pm = build_pm(routing_pass, layout_pass, cm)
```

```python
qc_tr = pm.run(qc_decomposed)
qc_tr.draw(output='mpl')
```
